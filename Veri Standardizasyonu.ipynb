### Veri Standardizasyonu

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3})

df = df.astype(float)
df
```

#### Standardizasyonu

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

```python
df.mean()
```

#### Normalizayon

```python
preprocessing.normalize(df)
```

#### Min-Max Dönüşümü

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
```

```python
scaler.fit_transform(df)
```

### Değişken Dönüşümleri

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

#### 0-1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

#### 1 ve Diğerleri Dönüşümü

```python
df.head()
```

```python
#Dayi Sun olanlara 1 sağlamayanlara 0
df["yeni_day"] = np.where(df["day"].str.contains("Sun"), 1, 0)
```

```python
df
```

#### Çok Sınıflı Dönüşüm

```python
lbe.fit_transform(df["day"])
```

#### Kategorik değişken nümerik değerlere dönüştürüldüğü zaman ortaya bir bozukluk ortaya çıkar bu problemi gidermek için One-Hot dönüşümü yapılır

### One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
df.head()
```

```python
#prefix oluşturulacak olan değişkenlerin ön isimlendirilmesinde kullanılır
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"])
```

```python
df_one_hot.head()
```

#### Dummy değişken tuzağı: Eğer veri içerisindeki değişkenlere bir dönüşüm uygulandığında eğer bu dönüşüm sonrasında oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa bu dummy değişken tuzağını oluşturur.
#### Kısaca bir değişkeni ifade eden başka bir değişken varsa bu duruma dummy değişken tuzağı denir 

```python
pd.get_dummies(df, columns = ["day"], prefix = ["day"]).head()
```
